```python
import pandas as pd

# load datasets
df = pd.read_csv("../data/ab_test_updated.csv")
```

```python
import numpy as np

# conversions
conversions = df.groupby('group')['converted'].sum()
n = df.groupby('group')['converted'].count()

# Beta distribution parameters
control_alpha = conversions['control'] + 1
control_beta = n['control'] - conversions['control'] + 1

treatment_alpha = conversions['treatment'] + 1
treatment_beta = n['treatment'] - conversions['treatment'] + 1

# sampling
control_samples = np.random.beta(control_alpha, control_beta, 100000)
treatment_samples = np.random.beta(treatment_alpha, treatment_beta, 100000)

# probability treatment > control
prob = (treatment_samples > control_samples).mean()

print("P(Treatment > Control):", prob)
```
