```python
import pandas as pd

# load datasets
df = pd.read_csv("../data/ab_test_updated.csv")
```

```python
from statsmodels.stats.proportion import proportions_ztest

# conversions
conversions = df.groupby('group')['converted'].sum()

# total users
n = df.groupby('group')['converted'].count()

z_stat, p_value = proportions_ztest(
    count=[conversions['treatment'], conversions['control']],
    nobs=[n['treatment'], n['control']],
    alternative='larger'   # one-sided test
)

print("Z-stat:", z_stat)
print("P-value:", p_value)
```

```python
import numpy as np

control_cr = conversions['control'] / n['control']
treatment_cr = conversions['treatment'] / n['treatment']

diff = treatment_cr - control_cr

# standard error
se = np.sqrt(
    (control_cr * (1 - control_cr)) / n['control'] +
    (treatment_cr * (1 - treatment_cr)) / n['treatment']
)

# 95% CI
z = 1.96

lower = diff - z * se
upper = diff + z * se

print("Difference:", diff)
print("95% CI:", (lower, upper))
```
